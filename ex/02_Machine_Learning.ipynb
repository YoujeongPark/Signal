# Membrne 

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score
from sklearn.metrics import mean_squared_error
```

```python
membrane = df.copy()
#membrane.drop(['path'+n + '_bending_SINT','dataLength'],axis=1)
```

```python
x_data = membrane.drop(['path'+n + '_bending_SINT','dataLength','path'+n + '_membrane_SINT'] , axis= 1)
y_data = membrane['path'+n + '_membrane_SINT']
```

```python
x_data
```

```python
# scaler = StandardScaler()
# scaler.fit(x_data)
```

```python
#xtrain, xtest, ytrain, ytest = train_test_split(x_data, y_data, test_size=0.2, random_state=0)
```

```python
lin_reg = LinearRegression()
scores = cross_val_score(lin_reg, x_data, y_data, cv=10, n_jobs=-1, scoring = "neg_mean_squared_error")
lin_rmse_scores = np.sqrt(-scores)
LR_result = lin_rmse_scores.mean()
LR_result
```
